```python
from google.colab import files
import zipfile
import os

# Upload the archive.zip file
uploaded = files.upload()

# Path to the uploaded file in the Colab environment
zip_path = 'archive.zip'  # The file will be available here after upload
extract_path = '/content/unzipped_data'

# Create the extraction directory if it doesn't exist
os.makedirs(extract_path, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# Check if the file is extracted
print(f"Files extracted to {extract_path}")
print(os.listdir(extract_path))  # List files in the extraction directory
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import os

class DocumentRetrieval:
    def __init__(self, documents):
        """Initialize with a list of documents"""
        self.documents = documents
        self.vectorizer = TfidfVectorizer(stop_words='english')
        self.tfidf_matrix = self.vectorizer.fit_transform(documents)

    def search(self, query, top_k=3):
        """Search for top_k most relevant documents for the query"""
        query_vector = self.vectorizer.transform([query])
        similarities = cosine_similarity(query_vector, self.tfidf_matrix)[0]
        top_indices = similarities.argsort()[-top_k:][::-1]

        results = []
        for idx in top_indices:
            results.append({
                'document_id': idx,
                'similarity': similarities[idx],
                'text': self.documents[idx][:200] + "..."  # Show first 200 chars
            })
        return results

def load_dataset(dataset_path):
    """Load text documents from the extracted dataset"""
    documents = []
    file_paths = []

    # Ensure the path exists
    if not os.path.exists(dataset_path):
        print(f"Error: Dataset directory '{dataset_path}' not found.")
        return [], []

    # Walk through the directory
    for root, _, files in os.walk(dataset_path):
        for file in files:
            # Adjust this condition based on your file types
            if file.endswith(('.txt', '.json')):  # You can add more types as needed
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r', encoding='utf-8') as f:
                        text = f.read()
                        documents.append(text)
                        file_paths.append(file_path)
                except Exception as e:
                    print(f"Error reading {file_path}: {e}")

    return documents, file_paths

def main():
    # Path to the extracted dataset (update if needed)
    dataset_path = '/content/unzipped_data'  # Update with actual location if different

    # Load the documents
    print("Loading documents...")
    documents, file_paths = load_dataset(dataset_path)

    if not documents:
        print("No documents found! Make sure the path to extracted files is correct and the files are not empty.")
        return

    print(f"Loaded {len(documents)} documents")
    print("Sample file paths:", file_paths[:3])  # Show first 3 file paths

    # Initialize the retrieval system
    retrieval_system = DocumentRetrieval(documents)

    # Example queries - modify these based on your dataset content
    queries = [
        "machine learning algorithms",
        "data preprocessing techniques",
        "neural networks architecture"
    ]

    # Allow user to input custom queries
    custom_query = input("\nEnter your search query (or press Enter to use example queries): ")
    if custom_query:
        queries = [custom_query]

    # Process each query
    for i, query in enumerate(queries, 1):
        print(f"\nQuery {i}: '{query}'")
        print("-" * 50)

        results = retrieval_system.search(query)

        for rank, result in enumerate(results, 1):
            print(f"\nRank {rank}")
            print(f"Document: {file_paths[result['document_id']]}")  # Show the file path
            print(f"Similarity Score: {result['similarity']:.4f}")
            print(f"Preview: {result['text']}")
            print("-" * 30)

if __name__ == "__main__":
    main()
```
